Q2. Why is request used in Flask ?

Ans.In flask , a web framework is used for Python, a " request" object is used to handle incomming HTTP requests from clients, such as browser or mobile apps .The "request" object contains information about the current request being made to thge server ,including data from the client like form inputs ,query parameters ,headers and more.

Using the request object in flask allows us to access and manipulate the data sent by tghe client ,enabling us to create dynamic and interactive web applications.eg.

(a).Accessing From Data: If a user submits a form on a web page , the form data can be accessed using the "request.form" attribute. This allows us retrieve user inputs and process them.

(b).Handling Query Parameters: If a user includes query parameters in the URL(eg.,example.com/search?q=x), we can "request.args" to access those parameters and perform actions based on them.

(c).Reading Headers: The "request.headers" attribute allows us to access HTTP headers sent by the client ,which can be useful for authentication ,user agent detection , and more.

(d).File Uploads: When users upload files ,the "request.files" attribute enables us to handle file uploads and save the uploaded files to the server .

Overall, the "request" object in flask provides us with a convenient way to interact with client data and make informed decissions based on the information sent with each HTTP request.

Q3. Why is redirect() used in Flask ?

Ans. In Flask ,the "redirect()" function is used to perform a URL redirection.URL redircton is the process of sending a user from one URL to another .This is a common practice in web development for various purposes, such as :

(a).Page Navigation: We can use "redirect()" to guide users to different page or route within our application .For example ,after a user successfully submits a form , we might redirect them to a "welocome" page.

(b).Handling Authentication: After a user logs in ,we can use redirection to send them to a secure page , ensuring thry have the have the necessary access rights.


(c).Clean URLs: Redirects can be used to ensure that users always see clean ,user-friendly URLs ,even if the actual routes are more complex.

(d).Changing URL Structure: If we need to cahnge the structure of our URLs or reorganize our applicatiion's routes ,redirects can help ensure that existing links or bookmarks remain functional.

The "render_template()" function in Flask is used to render these templates . By using 'render_template()'functionn ,we can keep our applicaton logic separate from the presentation logic ,majing our code organized and maintanable .It also allows for better code reusability and promotes the user of a modular architecture in our web application.
